```python
# install required libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

**Download and Load the Data**


```python
# download dataset 
!wget -O survey-data.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv
```

```python
df = pd.read_csv("survey-data.csv")

df.head()
```

```python
for col in df.columns:
    print(col)
```

### Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
plot_data = df['DatabaseWantToWorkWith'].dropna().str.split(';').sum()
db_counts = Counter(plot_data)
top5_db = db_counts.most_common(5)
 
print(db_counts)
print(top5_db)
```

```python
plt.figure(figsize=(6, 6))
labels, sizes = zip(*top5_db)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20b.colors[2:7])
plt.axis('equal')
plt.title('Top 5 Databases Developers Want to Work With')
plt.tight_layout()
plt.show()
```
